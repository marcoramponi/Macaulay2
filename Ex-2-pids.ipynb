# Explorations with Macaulay2

### Exercise 2

Compute the Groebner basis of the following polynomial ideals generated by 3 polynomials in $\mathbb{Q}[x]$ with respect to lex and other term orders. How many generators do they have? Explain your results.

---

```macaulay2
R = QQ[x];
```

```macaulay2
f1 = 4*x^4-4*x^3-41*x^2+65*x-42;
f2 = 12*x^4+44*x^3-11*x^2-43*x+70;
f3 = 6*x^4-17*x^3-12*x^2+37*x-30;
```

```macaulay2
I = ideal(f1,f2,f3);
```

```macaulay2
gs = gens gb I
```

---

We get only one generator. Let us try different monomial orders...

```macaulay2
R = QQ[x, MonomialOrder=>RevLex, Global=>false]
```

```macaulay2
I = ideal(f1,f2,f3);

gs = gens gb I
```

---

Same result as before! We could keep on experimenting with any other ideal and always get a single generator of the Groebner basis; and the same one no matter which monomial ordering we choose...

By reflecting shortly upon this, we get to the following:

#### Conclusion

* On the set $\left\{ x^n \mid n \in \mathbb{N} \right\}$, the only monomial orders are the natural ordering and its converse, the latter of which is not a well-ordering (existence of least element does not hold). Therefore, we would get the same result by trying any other pre-defined monomial order. 
* The Groebner basis will consist of a single element, providing us with a single generator for (any) ideal $I$, because $\mathbb{Q}[x]$ is a PID!
